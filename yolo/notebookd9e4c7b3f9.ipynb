```python

from ultralytics import YOLO
import os, glob, shutil
from PIL import Image
```

```python

INPUT_PATH = "/kaggle/input/asl-alphabet" 
TRAIN_DIR = f"{INPUT_PATH}/asl_alphabet_train/asl_alphabet_train"
TEST_DIR = f"{INPUT_PATH}/asl_alphabet_test/asl_alphabet_test"

WORK_DIR = "/kaggle/working/asl_dataset"
os.makedirs(WORK_DIR, exist_ok=True)
os.makedirs(f"{WORK_DIR}/images/train", exist_ok=True)
os.makedirs(f"{WORK_DIR}/labels/train", exist_ok=True)
os.makedirs(f"{WORK_DIR}/images/val", exist_ok=True)
os.makedirs(f"{WORK_DIR}/labels/val", exist_ok=True)

classes = sorted(os.listdir(TRAIN_DIR))
class_map = {cls: idx for idx, cls in enumerate(classes)}
print("✅ Classes detected:", class_map)
```

```python
from tqdm import tqdm

def create_yolo_label(label_path, class_id):
    with open(label_path, "w") as f:
        f.write(f"{class_id} 0.5 0.5 1.0 1.0\n")

for cls in classes:
    img_dir = f"{TRAIN_DIR}/{cls}"
    imgs = glob.glob(f"{img_dir}/*.jpg")
    split_idx = int(0.9 * len(imgs))
    train_imgs, val_imgs = imgs[:split_idx], imgs[split_idx:]
    
    print(f"\n📂 Processing class '{cls}' ({len(imgs)} images)...")
    
    for img_path in tqdm(train_imgs, desc=f"Train {cls}", unit="img"):
        fname = os.path.basename(img_path)
        out_img = f"{WORK_DIR}/images/train/{cls}_{fname}"
        out_lbl = f"{WORK_DIR}/labels/train/{cls}_{fname.replace('.jpg','.txt')}"
        shutil.copyfile(img_path, out_img)
        create_yolo_label(out_lbl, class_map[cls])
    
    for img_path in tqdm(val_imgs, desc=f"Val {cls}", unit="img"):
        fname = os.path.basename(img_path)
        out_img = f"{WORK_DIR}/images/val/{cls}_{fname}"
        out_lbl = f"{WORK_DIR}/labels/val/{cls}_{fname.replace('.jpg','.txt')}"
        shutil.copyfile(img_path, out_img)
        create_yolo_label(out_lbl, class_map[cls])


```

```python

# 4. Create data.yaml
yaml_content = f"""
train: {WORK_DIR}/images/train
val: {WORK_DIR}/images/val

nc: {len(classes)}
names: {classes}
"""

with open(f"{WORK_DIR}/data.yaml", "w") as f:
    f.write(yaml_content)


print(yaml_content)
```

```python
model = YOLO("yolov8n.pt")
model.train(
    data=f"{WORK_DIR}/data.yaml",
    epochs=100,        
    imgsz=224,        
    batch=32,         
    workers=8,         
    device=0,         
    name="yolo"
)
```

```python

# 6. Evaluate Model
metrics = model.val()
print("✅ Validation metrics:", metrics)


# 7. Run Inference on Test Images

sample_imgs = glob.glob(f"{TEST_DIR}/*.jpg")
print(f"Found {len(sample_imgs)} test images.")

results = model.predict(sample_imgs, save=True, imgsz=224)
print("✅ Inference complete. Check runs/detect/asl_yolo/predict folder.")


# 8. Save Best Model

```
